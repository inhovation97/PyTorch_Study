````python
import visdom
vis = visdom.Visdom()
````

## 텍스트 표기

````python
vis.text("hello world", env="main")
````

## 이미지

````python
import numpy as np
r=np.random.randn(3, 30, 400)
vis.image(r)
````

## 그래프 그리기

````python
Y_data = np.random.randn(5)
plt = vis.line(Y=Y_data)
````

````python
for i in range(100):
    X_append = [i]
    Y_append = np.random.randn(1)
    vis.line(
        X=X_append, 
        Y=Y_append, 
        win=plt,
        update="append" # visdom에 이렇게 업데이트 인자가 있음
    )
````
